## рҹ“Ҡ View Your Outputs

Your generated narrations are saved to Google Drive. View them below:

```python
# List generated outputs
import os
from IPython.display import Audio, display

output_dir = env_info['output_path']

if os.path.exists(output_dir):
    files = sorted([f for f in os.listdir(output_dir) if f.endswith('.wav')])

    if files:
        print(f"рҹ“Ӯ Found {len(files)} narration(s) in {output_dir}:\n")

        for f in files[-5:]:  # Show last 5
            filepath = os.path.join(output_dir, f)
            print(f"рҹ”Ҡ {f}")
            display(Audio(filepath))
            print()
    else:
        print("No narrations generated yet.")
else:
    print(f"Output directory not found: {output_dir}")
```
